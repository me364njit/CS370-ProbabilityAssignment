```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the dataset
dataset_url = "imports-85.csv"
df = pd.read_csv(dataset_url)

# Select the target variable and feature columns
target_variable_column = df['city-mpg']
feature_columns = df[['curb-weight', 'engine-size']]

# Convert selected columns to NumPy arrays
y = target_variable_column.values
X = feature_columns.values

# Standardize the feature values (mean=0, std=1)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Add a bias term (intercept) to the features
X_bias = np.c_[np.ones(X_scaled.shape[0]), X_scaled]

# Hyperparameters
learning_rate = 0.01
num_epochs = 1000
l2_lambda = 0.01  # Regularization parameter

# Initialize weights randomly
np.random.seed(0)
weights = np.random.randn(X_bias.shape[1])

# SGD training loop
for epoch in range(num_epochs):
    for i in range(X_bias.shape[0]):
        # Randomly select a data point
        random_index = np.random.randint(X_bias.shape[0])
        xi = X_bias[random_index]
        yi = y[random_index]

        # Compute the prediction
        prediction = np.dot(xi, weights)

        # Compute the gradient of the L2 regularized loss with respect to weights
        gradient = -2 * xi * (yi - prediction) + 2 * l2_lambda * weights

        # Update weights using SGD
        weights -= learning_rate * gradient

# The trained weights contain the coefficients for the linear model
intercept, weight_curb_weight, weight_engine_size = weights

# Predict MPG for a given curb weight and engine size
new_curb_weight = 2555  # Replace with your own value
new_engine_size = 127   # Replace with your own value

# Standardize the new input data
new_data = np.array([new_curb_weight, new_engine_size])
new_data_scaled = scaler.transform(new_data.reshape(1, -1))

# Add a bias term
new_data_bias = np.append(1, new_data_scaled)

# Make a prediction
predicted_mpg = np.dot(new_data_bias, weights)
print("Predicted MPG:", predicted_mpg)


```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Load the dataset
dataset_url = "imports-85.csv"
df = pd.read_csv(dataset_url)

# Select the target variable and feature columns
target_variable_column = df['city-mpg']
feature_columns = df[['curb-weight']]

# Convert selected columns to NumPy arrays
y = target_variable_column.values
X = feature_columns.values

# Normalize the feature (mean=0, std=1)
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_normalized = (X - X_mean) / X_std

# Add a bias term (intercept) to the features
X_b = np.c_[np.ones((len(X_normalized), 1)), X_normalized]

def plot_batch_gradient_descent(theta, eta):
    m = len(X_b)
    plt.plot(X, y, "b.")
    n_epochs = 1000
    n_shown = 20
    theta_path = []
    for epoch in range(n_epochs):
        if epoch < n_shown:
            y_predict = X_b @ theta
            color = mpl.colors.rgb2hex(plt.cm.OrRd(epoch / n_shown + 0.15))
            plt.plot(X, y_predict, linestyle="solid", color=color)
        gradients = 2 / m * X_b.T @ (X_b @ theta - y)
        theta = theta - eta * gradients
        theta_path.append(theta)
    plt.xlabel("$x_1$")
    plt.axis([0, max(X)[0] + 500, 0, max(y) + 10])
    plt.grid()
    plt.title(fr"$\eta = {eta}$")
    return theta_path

np.random.seed(42)
theta = np.random.randn(2, 1)  # random initialization

plt.figure(figsize=(10, 4))
plt.subplot(131)
plot_batch_gradient_descent(theta, eta=0.02)
plt.ylabel("$y$", rotation=0)
plt.subplot(132)
theta_path_bgd = plot_batch_gradient_descent(theta, eta=0.1)
plt.gca().axes.yaxis.set_ticklabels([])
plt.subplot(133)
plt.gca().axes.yaxis.set_ticklabels([])
plot_batch_gradient_descent(theta, eta=0.5)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

# Load the dataset
dataset_url = "imports-85.csv"
df = pd.read_csv(dataset_url)

# Select the target variable and feature columns
target_variable_column = df['city-mpg']
feature_columns = df[['curb-weight']]

# Convert selected columns to NumPy arrays
y = target_variable_column.values
X = feature_columns.values

# Normalize the feature (mean=0, std=1)
X_mean = np.mean(X, axis=0)
X_std = np.std(X, axis=0)
X_normalized = (X - X_mean) / X_std

# Add a bias term (intercept) to the features
X_b = np.c_[np.ones((len(X_normalized), 1)), X_normalized]

theta_path_sgd = []  # To store the path of theta in the parameter space

n_epochs = 50
t0, t1 = 5, 50  # Learning schedule hyperparameters

def learning_schedule(t):
    return t0 / (t + t1)

np.random.seed(42)
theta = np.random.randn(2, 1)  # Random initialization

n_shown = 20
plt.figure(figsize=(6, 4))

for epoch in range(n_epochs):
    for iteration in range(len(X_b)):

        if epoch == 0 and iteration < n_shown:
            y_predict = X_b @ theta
            color = mpl.colors.rgb2hex(plt.cm.OrRd(iteration / n_shown + 0.15))
            plt.plot(X, y_predict, color=color)

        random_index = np.random.randint(len(X_b))
        xi = X_b[random_index:random_index + 1]
        yi = y[random_index:random_index + 1]
        gradients = 2 * xi.T @ (xi @ theta - yi)  # For SGD, do not divide by m

        # Calculate the learning rate, handling the case where the denominator is zero
        try:
            eta = learning_schedule(epoch * len(X_b) + iteration)
        except ZeroDivisionError:
            eta = learning_schedule(1)

        theta = theta - eta * gradients
        theta_path_sgd.append(theta)

print("Final theta values:\n", theta)

plt.plot(X, y, "b.")
plt.xlabel("$x_1$")
plt.ylabel("$y$", rotation=0)
plt.axis([0, max(X)[0] + 500, 0, max(y) + 10])
plt.grid()
plt.title("Stochastic Gradient Descent")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler

# Load the dataset
dataset_url = "imports-85.csv"
df = pd.read_csv(dataset_url)

# Select the target variable and feature columns
target_variable_column = df['city-mpg']
feature_columns = df[['curb-weight']]

# Convert selected columns to NumPy arrays
y = target_variable_column.values
X = feature_columns.values

# Create a pipeline for polynomial regression
plt.figure(figsize=(6, 4))

for style, width, degree in (("r-+", 2, 1), ("b--", 2, 2), ("g-", 1, 40)):
    polybig_features = PolynomialFeatures(degree=degree, include_bias=False)
    std_scaler = StandardScaler()
    lin_reg = LinearRegression()
    polynomial_regression = make_pipeline(polybig_features, std_scaler, lin_reg)
    polynomial_regression.fit(X, y)
    
    # Generate new data for prediction
    X_new = np.linspace(X.min(), X.max(), 100).reshape(-1, 1)
    y_newbig = polynomial_regression.predict(X_new)
    
    label = f"{degree} degree{'s' if degree > 1 else ''}"
    plt.plot(X_new, y_newbig, style, label=label, linewidth=width)

plt.plot(X, y, "b.", linewidth=3)
plt.legend(loc="upper right")
plt.xlabel("Curb Weight")
plt.ylabel("City MPG", rotation=0)
plt.axis([1200, 4500, 0, 55])
plt.grid()
plt.title("Polynomial Regression on 'imports-85.csv'")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import Lasso, Ridge
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# Load the dataset
dataset_url = "imports-85.csv"
df = pd.read_csv(dataset_url)

# Select the target variable and feature columns
target_variable_column = df['city-mpg']
feature_columns = df[['curb-weight']]

# Convert selected columns to NumPy arrays
y = target_variable_column.values
X = feature_columns.values

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Create a meshgrid for theta1 and theta2
t1a, t1b, t2a, t2b = -1, 3, -1.5, 1.5
t1s = np.linspace(t1a, t1b, 500)
t2s = np.linspace(t2a, t2b, 500)
t1, t2 = np.meshgrid(t1s, t2s)

# Fit Lasso and Ridge regression models
lasso_model = Lasso()
ridge_model = Ridge()

lasso_model.fit(X_scaled, y)
ridge_model.fit(X_scaled, y)

# Compute the cost function for Lasso and Ridge
J_lasso = np.abs(lasso_model.predict(X_scaled) - y).mean()
J_ridge = ((ridge_model.predict(X_scaled) - y) ** 2).mean()

# Plot the results
plt.figure(figsize=(12, 5))

# Plot Lasso
plt.subplot(121)
plt.plot(X_scaled, y, "b.")
plt.xlabel("Scaled Curb Weight")
plt.ylabel("City MPG")
plt.title("Lasso Regression")
plt.plot(X_scaled, lasso_model.predict(X_scaled), "r-")
plt.text(0.5, 30, f"Cost: {J_lasso:.2f}", fontsize=12, color='r')
plt.grid(True)

# Plot Ridge
plt.subplot(122)
plt.plot(X_scaled, y, "b.")
plt.xlabel("Scaled Curb Weight")
plt.ylabel("City MPG")
plt.title("Ridge Regression")
plt.plot(X_scaled, ridge_model.predict(X_scaled), "g-")
plt.text(0.5, 30, f"Cost: {J_ridge:.2f}", fontsize=12, color='g')
plt.grid(True)

plt.tight_layout()
plt.show()
```
